```python
import numpy as np
```

```python
array = np.array([[10, 20, 30],
                  [40, 50, 60],
                  [70, 80, 90]])
array
print(array.ndim)
print(array.shape)
```

```python
# Buscar elemento na linha 1 e coluna 2
elemento = array[1,2]
elemento
```

```python
linha_0 = array[0]
linha_0
```

```python
# Acessando a coluna 1
coluna_1 = array[:, 1]
coluna_1
```

```python
# Fatiando para obter um subarray
# linha 0 a 1 | Coluna 1 a 2
subarray = array[0:2, 1:3]
subarray
```

```python
mascara = array > 50
mascara
```

```python
elementos_maiores_que_50 = array[mascara]
elementos_maiores_que_50
```

```python
# Modificando elemento na linha 1, coluna 1
array[1, 1] = 100
array
```
